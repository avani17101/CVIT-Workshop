```python
path = 'images/'
from os import listdir
from os.path import isfile, join
files = [f for f in listdir(path) if isfile(join(path, f))]
```

```python
print(files)
```

```python
import numpy as np
files = np.array(files)
files.shape[0]
```

```python
#using 5 istead of 50!
import random
#filenos = np.random.randint(low=1,high=files.shape[0],size=5)
"""Note: above function returns dublicate numbers too hence we userandom.sample which doesnt return duplicate/repeating numbers"""
def load_data(k):
    """
    returns k randomly selected samples from dataset
    """
    filenos = random.sample(range(files.shape[0]), k)
    return filenos
```

```python
filenos = load_data(5)
```

```python
import matplotlib.pyplot as plt
images = []
import cv2
for n in filenos:
    image = cv2.imread(path+files[n],cv2.IMREAD_UNCHANGED)
    images.append(cv2.resize(image, (100,100)))
```

```python
fig = plt.figure(figsize=(50, 50))
for i in range(5):
    sub = fig.add_subplot(15, 1, i + 1)
    sub.imshow(images[i], interpolation='nearest')
```
